```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load the dataset
train_df = pd.read_csv('/kaggle/input/nlp-getting-started/train.csv')

test_df = pd.read_csv('/kaggle/input/nlp-getting-started/test.csv')
```

******Data Exploration & Pre-processing

```python
train_df.head()
```

```python
test_df.head()
```

```python
import re

def clean_text(text):
    text.lower() # convert to lowercase
    text = re.sub('http\S+|www.\S+', '', text)   # remove URLs
    text = re.sub('[^a-z]+', ' ', text)   # remove non-alphanumeric characters
    return text
```

```python
train_df['text'] = train_df['text'].apply(clean_text)
test_df['text'] = test_df['text'].apply(clean_text)
```

Feature Engineering

```python
from sklearn.feature_extraction.text import TfidfVectorizer

Tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7, max_features=5000)
X_train_tf = Tfidf_vectorizer.fit_transform(train_df['text'])
X_test_tf = Tfidf_vectorizer.transform(test_df['text'])
y_train = train_df['target']
y_test = test_df['id']
```

Model Training

Using a simple logistic regression model.

```python
from sklearn.linear_model import LogisticRegression


clf = LogisticRegression()
clf.fit(X_train_tf, y_train)
```

Model Evaluation on Training Data

check the performance on the training data first

```python
from sklearn.metrics import accuracy_score, classification_report

y_train_predict = clf.predict(X_train_tf)
print("Training Accuracy:", accuracy_score(y_train, y_train_predict))


print("Classification report:", classification_report(y_train, y_train_predict))
```

```python
y_test_predict = clf.predict(X_test_tf)
```

```python


submission = pd.DataFrame({
    "id": test_df['id'],
    "target": y_test_predict
})

submission.to_csv('submission.csv', index=False)




```
